```python
import pandas as pd
```

```python
data = pd.read_csv("prem.csv")
data.head()
```

```python
data.info()
```

Split into 2 dataframes: one for all home games, the other for all away games.

```python
home_statistics = data[["home_team", "home_goals", "home_xG", "home_posession", "home_passing_accuracy", 
                             "home_shots_on_target", "home_shots", "home_saves", "home_fouls", "home_corners", 
                             "home_crosses", "home_touches", "home_tackles", "home_interceptions", 
                              "home_aerials_won", "home_clearances", "home_offsides", "home_goal_kicks", 
                             "home_throw_ins", "home_long_balls"]].copy()
home_statistics.head()
```

```python
away_statistics = data[["away_team", "away_goals", "away_xG", "away_possession", "away_passing_accuracy", 
                             "away_shots_on_target", "away_shots", "away_saves", "away_fouls", "away_corners", 
                             "away_crosses", "away_touches", "away_tackles", "away_interceptions", 
                              "away_aerials_won", "away_clearances", "away_offsides", "away_goal_kicks", 
                              "away_throw_ins", "away_long_balls"]].copy()
away_statistics.head()
```

Standardize all column names in both data frames to concatenate home data with away data. 

```python
home_statistics.columns = ["team", "goals", "xG", "possession", "passing_accuracy", "shots_on_target", "shots", 
                           "saves", "fouls", "corners", "crosses", "touches", "tackles", "interceptions", 
                           "aerials_won", "clearances", "offsides", "goal_kicks", "throw_ins", "long_balls"]
home_statistics.head()
```

```python
away_statistics.columns = ["team", "goals", "xG", "possession", "passing_accuracy", "shots_on_target", "shots", 
                           "saves", "fouls", "corners", "crosses", "touches", "tackles", "interceptions", 
                           "aerials_won", "clearances", "offsides", "goal_kicks", "throw_ins", "long_balls"]
away_statistics.head()
```

Concatenate home_statistics and away_statistics to aggregate all football match data. 

```python
all_statistics = pd.concat([home_statistics, away_statistics], ignore_index = True)
all_statistics.head()
```

```python
all_statistics.info()
```

The "possession" and "passing_accuracy" columns have data type "object". We will express the percentages by converting the strings to floating point numbers. 

```python
all_statistics["possession"] = all_statistics["possession"].str.rstrip("%").astype("float") / 100.0
all_statistics["passing_accuracy"] = all_statistics["passing_accuracy"].str.rstrip("%").astype("float") / 100.0
all_statistics.head()
```

```python
all_statistics.info()
```

```python
all_statistics.to_csv("stats1.csv", index=False)
```
